```python
import numpy as np
```

```python
#skiprows=1 == pular primeira linha
#skip_header=1 == pular cabeÃ§alho
#unpack=True == alocar colunas em valores ex:(col1 => val1)
#filling_values=-999 == preencher valor faltantes
```

```python
val1, val2, val3 = np.loadtxt('dados.txt', skiprows=1, unpack=True)
print(val1)
print(val2)
print(val3)
```

```python
my_array = np.genfromtxt('dados2.txt', skip_header=1, filling_values=-999)
print(my_array)
```
